## Visualizations

```python
import matplotlib.pyplot as plt
# for plots
import seaborn
from mpl_toolkits import mplot3d
import numpy as np

seaborn.set()
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}

pylab.rcParams.update(params)
```

```python
def c(x, y, s):
    return (1 - x*y)**2 + (s)*(x**2) + (s)*(y**2)

x = np.linspace(-1.2, 1.2, 30)
y = np.linspace(-1.2, 1.2, 30)

sig = 0
X, Y = np.meshgrid(x, y)
Z = c(X, Y, sig)
```

```python
seaborn.heatmap(Z, annot=True, cmap="YlGnBu")
plt.title("2D representation of Potential Function, sigma = " + str(sig) )
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (20, 20))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 500, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.title("3D representation of Potential Function, sigma = " + str(sig) )
```
