```python
import matplotlib
import matplotlib.pyplot as plt
# for plots
import seaborn
from mpl_toolkits import mplot3d
import numpy as np

seaborn.set()
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}

pylab.rcParams.update(params)
```

```python
def c(x, y, s):
    b = 0
    return (1 - x*(y+b))**2 + (s)*(x**2) + (s)*(y**2)

x = np.linspace(-1.2, 1.2, 80)
y = np.linspace(-1.2, 1.2, 80)

sig = 0.5
X, Y = np.meshgrid(x, y)
Z = c(X, Y, sig)
```

```python
lm_array = np.linspace(0.1,0.9, 9)
targ = 1
```

```python
re_array = np.array([0.010016577285056267, 
                     0.04003564647494689, 
                     0.08989195718172907, 
                     0.1598749867358406, 
                     0.24990919667443992, 
                     0.36011960577084046, 
                     0.48895851887407654, 
                     0.6327276660030635, 
                     0.7793345019819279])

bc_array = np.array([0.10003290005989218, 
                     0.20007441578959737, 
                     0.29997870362513124, 
                     0.3999458427083926, 
                     0.49991860957500195, 
                     0.6000100402352508,
                     0.6994886846096227, 
                     0.7964266722943617, 
                     0.8849011517797951])
                    
dc_array = np.array([0.09999751390013233,
                     0.19997313571941178,
                     0.29992498509543286,
                     0.39993979358405973, 
                     0.5000029669740071, 
                     0.600121488549989, 
                     0.699483055105544, 
                     0.7963358856578288, 
                     0.8847474455767884])  
```

```python
re_expect = np.zeros(np.shape(lm_array))
c_expect = np.zeros(np.shape(lm_array))

for ia in range(len(lm_array)):
    re_expect[ia] = (lm_array[ia]**2)/(targ**2)
    c_expect[ia] = lm_array[ia]
```

```python
seaborn.set_style("whitegrid", {'axes.grid' : False})

fz = '40'
fz2 = '30'
plt.plot(lm_array, re_expect, '.--', linewidth=4, markersize=10, color='lightgray', label="$\epsilon$ (Anl.)")
plt.plot(lm_array, re_array, 'o', markersize=20, color='dimgray', label="$\epsilon$ (Sim.)")

plt.plot(lm_array, c_expect, '.--', linewidth=4, markersize=10, color='lightsteelblue', label="$c_B, c_D$ (Anl.)")
plt.plot(lm_array, bc_array, 'X', markersize=25, color='darkblue', label=" $c_B$ (Sim.)")
plt.plot(lm_array, dc_array, 'P', markersize=25, color='coral', label="$c_D$ (Sim.)")

plt.xlabel("Penalty Parameter ($\lambda$)", fontsize=fz)
plt.xticks(fontsize=fz2)
plt.ylabel("Stationary Cost and Task Error", fontsize=fz)
plt.yticks(fontsize=fz2)
# plt.legend(fontsize=24, bbox_to_anchor=(0.55, 0.37), loc='upper left')
plt.legend(fontsize=30)

# plt.title("Stationary Cost and Reach Error As a Function of $\lambda$", fontsize=fz)
print("Stationary Cost and Reach Error As a Function of $\lambda$")
```
