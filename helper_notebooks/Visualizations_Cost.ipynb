## Visualizations

```python
import matplotlib.pyplot as plt
# for plots
import seaborn
from mpl_toolkits import mplot3d
import numpy as np

seaborn.set()
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}

pylab.rcParams.update(params)
```

```python
def c(x, y, s):
    b = 0
    return (1 - x*(y+b))**2 + (s)*(x**2) + (s)*(y**2)

x = np.linspace(-1.2, 1.2, 80)
y = np.linspace(-1.2, 1.2, 80)

sig = 0.5
X, Y = np.meshgrid(x, y)
Z = c(X, Y, sig)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (20, 20))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 1000)
ax.set_xlabel('W', fontsize=50)
ax.set_ylabel('K', fontsize=50)
ax.set_zlabel('$\phi$');
plt.title("Potential Function, $\lambda$ = " + str(sig), fontsize=50)
ax.view_init(azim=-65)
```
